```python
import pandas as pd
import numpy as np
import sys
```

# Caso 1: Identificando flores

Dataset: Iris, disponível em http://archive.ics.uci.edu/ml/datasets/Iris

Problema: identificar uma flor utilizando as medições da pétala e da sépala

```python
# lendo os dados a partir de um arquivo csv
data = pd.read_csv('./data/iris.data', delimiter=',', header=None,
                   names=['sepal length', 'sepal width', 'petal length', 'petal width', 'class'])
data.head()
```

```python
np.unique(data['class'])
```

# Separando os dados
![Separação em treino x teste](./images/train-test.png)
Fonte: https://jessesw.com/Rec-System/

```python
setosa = data[data['class'] == 'Iris-setosa']
versicolor = data[data['class'] == 'Iris-versicolor']
virginica = data[data['class'] == 'Iris-virginica']

print(len(setosa), len(versicolor), len(virginica))
```

```python
train = pd.concat([setosa.iloc[:30], versicolor.iloc[:30], virginica.iloc[:30]])
test = pd.concat([setosa.iloc[30:], versicolor.iloc[30:], virginica.iloc[30:]])

print(len(train), len(test))
```

```python
train_x = train.iloc[:, :4]
test_x = test.iloc[:, :4]

train_x.head()
```

```python
train_y = train.iloc[:, 4]
test_y = test.iloc[:, 4]

train_y.head()
```

# Nearest neighbor: O primeiro modelo/algoritmo de machine learning

![Nearest neighbor](./images/knn.png)
Fonte:https://www.python-course.eu/k_nearest_neighbor_classifier.php

```python
def distance(p1, p2):
    return np.sum((p1 - p2) ** 2)


def classify(sample, data, labels):
    min_d = sys.float_info.max
    min_d_index = -1
    for i in range(len(data)):
        train_sample = data.iloc[i]
        d = distance(sample, train_sample)
        if d < min_d:
            min_d = d
            min_d_index = i
    
    return labels[min_d_index]
```

```python
classify(test_x.iloc[0], train_x, train_y.iloc)
```

```python
predictions = []
for i in range(len(test_x)):
    p = classify(test_x.iloc[i], train_x, train_y.iloc)
    predictions.append(p)
```

```python
print(predictions)
```

```python
evaluations = [1.0 if y == y_hat else 0.0 for (y, y_hat) in zip(test_y, predictions)]
accuracy = sum(evaluations) / float(len(evaluations))
print(accuracy)
```

# Caso 2: detecção de câncer

Dataset: Breast Cancer Wisconsin (Diagnostic) Data Set, disponível em http://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+%28Diagnostic%29

Problema: Identificar se o câncer é maligno ou benigno

```python
data = pd.read_csv('./data/wbcd.data', delimiter=',', header=None)
data.head()
```

```python
data['label'] = 0
data.loc[data[1] == 'B', 'label'] = 1
data.head(20)
```

```python
y = data['label'].as_matrix()
y
```

```python
(RADIUS, PERIMETER, AREA) = (2, 4, 5)
X = data[[RADIUS, PERIMETER, AREA]].as_matrix()
X
```

```python
indices = np.arange(len(y))
np.random.shuffle(indices)
train_indices = indices[:400]
test_indices = indices[400:]
```

```python
train_x = X[train_indices]
train_y = y[train_indices]

test_x = X[test_indices]
test_y = y[test_indices]
```

```python
import autograd.numpy as np
from autograd import grad
```

```python
# transforma data frame em matriz
(M, N) = train_x.shape

C = 1 # número de classes
W = np.random.normal(size=(N, C), loc=0, scale=0.01)
b = np.zeros(shape=(M, C))

def f(X, W, b):
    h = np.dot(X, W) + b
    e = np.exp(h)
    return e / (e + 1)

f(train_x, W, b)

df_dx = grad(f)
```
