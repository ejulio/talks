```
import tensorflow as tf
import numpy as np
```

```
i = tf.keras.layers.Input(shape=(2,))

l = tf.keras.layers.Dense(2,
                          activation='tanh',
                          kernel_initializer='random_normal',
                          bias_initializer='zeros',
                          use_bias=True,
                          bias_regularizer=None,
                          kernel_regularizer=None)(i)

o = tf.keras.layers.Dense(1,
                          activation='sigmoid',
                          kernel_initializer='random_normal',
                          bias_initializer='zeros',
                          use_bias=True,
                          bias_regularizer=None,
                          kernel_regularizer=None)(l)

model = tf.keras.models.Model(inputs=i, outputs=o)

sgd = tf.keras.optimizers.SGD(lr=0.1, momentum=0.9, decay=1e-6)
model.compile(optimizer=sgd,
              loss='mean_squared_error',
              metrics=['accuracy'])

x = np.array([
    [0, 0],
    [1, 0],
    [0, 1],
    [1, 1]
]).astype('float32')

y = np.array([
    [0],
    [1],
    [1],
    [0]
]).astype('float32')

model.fit(x, y, epochs=1000, batch_size=4)
```
